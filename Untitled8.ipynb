```python
import numpy as np
import pymc as pm
import arviz as az

# --- Step 1: Generate or load data ---
np.random.seed(42)
observed_data = np.random.normal(loc=10, scale=1.0, size=100)  # e.g. 100 samples from N(10, 1)

# --- Step 2: Build the model ---
with pm.Model() as model1:
    # Prior for theta (mean)
    theta = pm.Normal("theta", mu=0, sigma=5)  
    
    # Prior for tau (precision)
    # Gamma is often parameterized by alpha/beta (a, b), or shape/rate.
    # In PyMC, pm.Gamma("name", alpha, beta) uses alpha=shape, beta=rate
    tau = pm.Gamma("tau", alpha=2.0, beta=1.0)
    
    # Likelihood: Normal with mean=theta, sd=sigma = 1/sqrt(tau)
    # Alternatively you can directly model precision=1/sigma^2 in pm.Normal(..., tau=tau)
    sigma = pm.Deterministic("sigma", 1.0 / pm.math.sqrt(tau))
    
    obs = pm.Normal("obs", mu=theta, sigma=sigma, observed=observed_data)
    
    # --- Step 3: Sample from the posterior ---
    trace1 = pm.sample(
        draws=2000,      # total MCMC samples
        tune=1000,       # warm-up / tuning steps
        target_accept=0.95,
        chains=4,
        random_seed=42
    )

# --- Step 4: Diagnostics ---
az.summary(trace1, var_names=["theta", "tau", "sigma"])
az.plot_trace(trace1, var_names=["theta", "tau", "sigma"])
az.plot_pair(trace1, var_names=["theta", "tau"])
```

```python
with pm.Model() as model2:
    # e.g. a StudentT prior for theta
    # StudentT(nu, mu, sigma)
    theta = pm.StudentT("theta", nu=3, mu=0, sigma=5)
    
    # a HalfNormal prior for tau (precision)
    # often used for scale parameters, but we can adapt it for precision
    # Alternatively, you might want to invert it carefully for use with pm.Normal(..., tau=...)
    tau = pm.HalfNormal("tau", sigma=2.0)  # e.g. half-normal with std=2
    
    # As before, define sigma = 1/sqrt(tau)
    sigma = pm.Deterministic("sigma", 1.0 / pm.math.sqrt(tau))
    
    obs = pm.Normal("obs", mu=theta, sigma=sigma, observed=observed_data)
    
    trace2 = pm.sample(
        draws=2000, tune=1000, target_accept=0.95, chains=4, random_seed=42
    )
    
az.summary(trace2, var_names=["theta", "tau", "sigma"])
az.plot_trace(trace2)
az.plot_pair(trace2)
```

```python
with pm.Model() as model3:
    # Laplace(mean, b) with e.g. mean=0, b=2
    theta = pm.Laplace("theta", mu=0, b=2)
    
    # Exponential prior for tau with lambda=1
    tau = pm.Exponential("tau", lam=1.0)
    
    sigma = pm.Deterministic("sigma", 1.0/pm.math.sqrt(tau))
    
    obs = pm.Normal("obs", mu=theta, sigma=sigma, observed=observed_data)
    
    trace3 = pm.sample(
        draws=2000, tune=1000, target_accept=0.95, chains=4, random_seed=42
    )

az.summary(trace3, var_names=["theta", "tau", "sigma"])
az.plot_trace(trace3)
az.plot_pair(trace3)
```

```python
###Q3
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# --------------------------------------------
# 1) Define the target density function
#    (here we use Beta(2,5) as an example).
#    The function returns 0 outside [0,1].
# --------------------------------------------
def target_pdf(x):
    if x < 0 or x > 1:
        return 0.0
    return stats.beta(a=2, b=5).pdf(x)

# --------------------------------------------
# 2) Implement a simple 1D slice sampling function
#    - x0: current sample point
#    - pdf: the (unnormalized) target density
#    - step_out: initial step size for bracketing the slice
#    - max_step_out: maximum expansions to avoid infinite loops
# --------------------------------------------
def slice_sample_1d(x0, pdf, step_out=0.1, max_step_out=50, random_state=None):
    rng = np.random.default_rng(random_state)
    
    # 1) Draw y ~ Uniform(0, f(x0))
    fx0 = pdf(x0)
    y = rng.uniform(0, fx0)
    
    # 2) Find the slice interval [left, right].
    #    Start by positioning left near x0, then expand step_out as needed.
    left = x0 - step_out * rng.random()  # shift slightly to avoid always symmetrical bracketing
    right = left + step_out
    
    if left > right:
        left, right = right, left

    # Step out to the left
    i = 0
    while pdf(left) > y and i < max_step_out:
        left -= step_out
        i += 1
        if left < 0:  # For Beta, the domain is [0,1], so clamp at 0
            left = 0
            break
    
    # Step out to the right
    i = 0
    while pdf(right) > y and i < max_step_out:
        right += step_out
        i += 1
        if right > 1:
            right = 1
            break

    # 3) Now repeatedly sample within [left, right].
    #    If the chosen x_new does not satisfy f(x_new) >= y,
    #    shrink the interval around x_new, and repeat.
    while True:
        x_new = rng.uniform(left, right)
        if pdf(x_new) >= y:
            return x_new
        else:
            if x_new < x0:
                left = x_new
            else:
                right = x_new

# --------------------------------------------
# 3) Perform slice sampling and plot results
# --------------------------------------------
n_samples = 2000
samples = np.zeros(n_samples)
samples[0] = 0.5  # initial guess

for t in range(n_samples - 1):
    samples[t+1] = slice_sample_1d(samples[t], target_pdf)

# 4) Visualization: compare histogram of samples vs. the Beta(2,5) PDF
x_grid = np.linspace(0, 1, 200)
pdf_values = [target_pdf(xx) for xx in x_grid]

plt.figure(figsize=(8,4))
# Plot histogram of samples
plt.hist(samples, bins=30, density=True, alpha=0.5, label="Slice Samples")
# Plot theoretical Beta(2,5) PDF
plt.plot(x_grid, pdf_values, 'r-', linewidth=2, label="Beta(2,5) PDF")

plt.title("Slice Sampling for Beta(2,5)")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.show()
```
